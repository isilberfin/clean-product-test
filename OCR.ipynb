```python
import keras_ocr
import matplotlib.pyplot as plt
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
# Read images from folder path to image object
images = [
    keras_ocr.tools.read(img) for img in ['including.jpeg']
]
```

```python
# Load the images
images = [plt.imread('including.jpeg')]

# Get the text predictions for each image
prediction_groups = pipeline.recognize(images)

# Plot the text predictions for all images
fig, ax = plt.subplots(figsize=(10, 20))
for image, predictions in zip(images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)

plt.show()
```
